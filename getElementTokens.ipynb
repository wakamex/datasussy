```python
import requests
from datetime import datetime, timedelta, date
import pandas as pd
# contract_url_old = 'https://raw.githubusercontent.com/element-fi/elf-deploy/8a8a92b79622edfa35c1341859f7b125ab69e37b/addresses/mainnet.json'
# element_token_mappings_old = requests.get(contract_url_old).json()
contract_url = 'https://raw.githubusercontent.com/element-fi/elf-deploy/main/addresses/mainnet.json'
element_token_mappings = requests.get(contract_url).json()
```

```python
tokens = []
# display(element_token_mappings)
for token, data in element_token_mappings['tranches'].items():
    for tranche in data:
        # display(tranche)
        tokens.append({
            'expiry_timestamp': tranche['expiration'],
            'expiry_datetime': datetime.fromtimestamp(tranche['expiration']).strftime('%Y-%m-%d %H:%M:%S'),
            'token_address_raw': '\\'+tranche['address'][1:].lower(),
            'token':token,
            'token_type': 'pToken'
        })
        for token_type in ('ptPool', 'ytPool'):
            if token_type in tranche:
                tokens.append({
                    'expiry_timestamp': tranche['expiration'],
                    'expiry_datetime': datetime.fromtimestamp(tranche['expiration']).strftime('%Y-%m-%d %H:%M:%S'),
                    'short_date': datetime.fromtimestamp(tranche['expiration']).strftime('%d%b%Y'),
                    'token_address_raw': '\\'+tranche[token_type]['address'][1:].lower(),
                    'token':token,
                    'token_type': token_type,
                    'poolId': '\\'+tranche[token_type]['poolId'][1:].lower()
                })
tokens = pd.DataFrame(tokens)
```

```python
# display(tokens.loc[(tokens.token=='alusd3crv-f') & (tokens.token_type=='pToken'),:])
to_display=tokens.loc[(tokens.token_type=='pToken'),:].reset_index(drop=True)
print(f"all pTokens, to count # of pools = {len(to_display)}")
display(to_display)
```

```python
wrappedTokens = []
# display(element_token_mappings['wrappedPositions'])
for version in element_token_mappings['wrappedPositions'].keys():
    for wrapperType,data in element_token_mappings['wrappedPositions'][version].items():
        for token,address in data.items():
            if wrapperType=='yearnv4':
                wrappedTokens.append((version,token+'v4',address))
            else:
                wrappedTokens.append((version,token,address))
wrappedTokens = pd.DataFrame(wrappedTokens,columns={'version','token','address'})
display(wrappedTokens.style.hide_index())
```

```python
vaults = []
for vaultType,data in element_token_mappings['vaults'].items():
    for vault,address in data.items():
        if vaultType=='yearnv4':
            vaults.append((vault+'v4',address))
        else:
            vaults.append((vault,address))
vaults = pd.DataFrame(vaults,columns={'vault','address'})
display(vaults.style.hide_index())
```
